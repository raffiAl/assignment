```python
# ALAT"

import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.linear_model import Perceptron
```

```python
# BACA DATA

data = pd.read_excel("skorujian.xlsx")
x = data.drop('hasil', axis=1)
y = data['hasil']

x.head()
```

```python
# PREDIKSI DATA

clf = Perceptron(max_iter=5, tol=0, random_state=0)
clf.fit(x, y)

prediksi = clf.predict(x)
print(prediksi)
print(y)
```

```python
# AKURASI DATA

from sklearn.metrics import accuracy_score

akurasi = accuracy_score(y, prediksi)
print(f'Akurasi Model: {akurasi * 100}%')

from sklearn.metrics import classification_report
print('\nLaporan Klasifikasi:')
print(classification_report(y, prediksi))
```

```python
# TEST

test = np.array([0.7, 0.6]).reshape(1, -1)
print(clf.predict(test))
```

```python
# VISUALISASI SLP

import matplotlib.pyplot as plt

def visual_slp(input_size, output_size):
  fig, ax = plt.subplots(figsize=(10, 10))

  for i in range(input_size):
    neuron_x = 1
    neuron_y = (i + 1)/(input_size + 1)
    ax.scatter(neuron_x, neuron_y, color='blue', s=300)
    ax.text(neuron_x, neuron_y, f'Input\nNeuron {i+1}', ha='center', va='center', fontsize=10)


  for i in range(output_size):
    neuron_x = 2
    neuron_y = (i + 1)/(output_size + 1)
    ax.scatter(neuron_x, neuron_y, color='red', s=300)
    ax.text(neuron_x, neuron_y, f'Output\nNeuron {i + 1}', ha='center', va='center', fontsize=10)

  for i in range(input_size):
    for j in range(output_size):
      inputN_x = 1
      inputN_y = (i + 1)/(input_size + 1)
      outputN_x = 2
      outputN_y = (j + 1)/(output_size + 1)
      ax.plot([inputN_x, outputN_x], [inputN_y, outputN_y], color='black', linewidth=1)

  ax.set_title('Visualisasi single layer perceptron')
  ax.set_xlabel(f'Kesimpulan SLP: Model ini menggunakan arsitektur Single\nLayer Perceptron dengan 2 neuron input (fitur) yang\n terhubung langsung ke 1 neuron output (prediksi) tanpa lapisan tersembunyi (hidden layer)')
  ax.set_ylabel('Neuron')
  ax.set_xlim(0.5, 2.5)
  ax.set_ylim(0, 1)
  ax.grid(True)
  plt.xticks([1, 2], ['Input', 'Output'])
  plt.show()

ipS = 2
opS = 1
visual_slp(ipS, opS)
```
